#Importing Libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
```

Uploading Files

```python
uploaded = files.upload()
```

Load and Check the Dataset

```python
file_name = list(uploaded.keys())[0]
data = pd.read_csv(file_name)
```

#Check the missing Values

```python
print(data.isnull().sum())
```

# Fill missing 'Age' with median

```python
data.fillna({'Age': data['Age'].median()}, inplace=True)
```

# Fill missing 'Embarked' with mode

```python
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])
```

# --- Univariate Analysis ---

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Pclass', data=data)
plt.title('Passenger Class Distribution')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', data=data)
plt.title('Gender Distribution')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(data['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

# --- Bivariate Analysis ---

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Pclass', y='Age', data=data)
plt.title('Age Distribution Across Classes')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Embarked', y='Fare', data=data)
plt.title('Average Fare by Embarkation Port')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Age', y='Fare', hue='Pclass', data=data)
plt.title('Fare vs Age colored by Pclass')
plt.show()
```
